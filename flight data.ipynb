```python
import matplotlib.pyplot as plt

month=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
flight_1=[112,118,132,129,121,135,148,148,139,119,104,118]
flight_2=[88,100,65,170,96,100,160,50,99,126,140,144]
```

```python
plt.plot(month, flight_1, 'r>--')
plt.plot(month,flight_2, 'bo-')
plt.title("flight per month", fontsize = 20)
plt.xlabel("month" , fontsize = 15)
plt.ylabel("flight" , fontsize = 15)
plt.legend(['flight_1','flight_2'])
loc=10
plt.show()
```
